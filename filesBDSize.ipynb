```python
import os
import math
```

```python
# Dir onde os ficheiros de áudio estão
#dir = "/mnt/dataset/public/Fingerprinting/neural-audio-fp-dataset/music/test-query-db-500-30s/query/snr_00dB_10dB_1s/"
#dir = "/mnt/dataset/public/Fingerprinting/neural-audio-fp-dataset/music/test-query-db-500-30s/db/"
dir = "/mnt/dataset/public/Fingerprinting/neural-audio-fp-dataset/music/test-dummy-db-100k-full/fma_full/"

# Lista para armazenar length dos áudio, lenght está em bytes
audioLength = []
```

```python
#calcular o length de cada áudio e armazenar essas lengths. No final um print de quantos áudios com x size existem.
def calculate_audio_length(caminho_arquivo):
    tamanho = os.path.getsize(caminho_arquivo)
    audioLength.append(tamanho)
```

```python
# Percorrer recursivamente todas as subpastas e o que está dentro
#-db
#   -000
#       -000003.wav
#       -...
#   -001
#   -...
#   -155
for folder, subfolder, files in os.walk(dir):
    for file in files:
        # Verificar se o arquivo é um arquivo de áudio (por exemplo, com extensão .wav)
        if file.endswith(".wav"):
            # Construir o caminho completo do arquivo
            filename = os.path.join(folder, file)
            # Calcular o tamanho do arquivo de áudio e armazenar na lista
            calculate_audio_length(filename)
```

```python
# Contar quantos elementos são iguais numa lista
def count_equal_elements(lista, elemento):
    counter = lista.count(elemento)
    return counter

lengthToSearch = 480044

# Conta os áudios com o tamanho pedido
quantidade_audios_com_tamanho = count_equal_elements(audioLength, lengthToSearch)

print(f"Existem {quantidade_audios_com_tamanho} áudios com tamanho {lengthToSearch} bytes.")
```

```python
# Função para contar quantas vezes cada valor único aparece numa lista
def count_unique_values(lista):
    counter = {}
    for value in lista:
        if value in counter:
            counter[value] += 1
        else:
            counter[value] = 1
    return counter

# Calcular a contagem de cada valor único nos tamanhos dos áudios
valuesCounted = count_unique_values(audioLength)

# quantas vezes cada valor
for value, counter in valuesCounted.items():
    print(f"Contém {counter} elementos de valor {value}.")
```

```python
# Função que calcula o número de segmentos presente em cada ficheiro de áudio
def count_segments(lista):
    segmentsList = [math.floor((((value - 44) / 2)-8000)/4000) + 1 for value in lista]
    return segmentsList

numberSegments = count_segments(audioLength)
```
