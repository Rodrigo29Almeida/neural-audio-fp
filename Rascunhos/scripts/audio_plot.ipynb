```python
audiopath='/mnt/dataset/public/Fingerprinting/neural-audio-fp-dataset/music/train-10k-30s/fma_small_8k_plus_medium_2k/000/000625.wav'
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Carregar o arquivo de áudio
frequencia_amostragem, amostras = wavfile.read(audiopath)

# Calcular a duração total do sinal de áudio
duracao = len(amostras) / frequencia_amostragem

# Criar vetor de tempo
tempo = np.linspace(0, duracao, len(amostras))

# Plotar o sinal de áudio no domínio do tempo com maior qualidade
plt.figure(figsize=(16, 8))  # Define o tamanho da figura
plt.plot(tempo, amostras)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal de Áudio no Domínio do Tempo')
plt.tight_layout()  # Ajusta automaticamente a disposição dos elementos na figura
plt.savefig('sinal_de_audio.png', dpi=600)  # Salva a figura em alta qualidade com DPI de 600
plt.show()
```
